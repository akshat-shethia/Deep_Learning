```python
X = tf.range(-100, 100, 4)
y = X + 10

X, y
```

```python
plt.scatter(X, y, color = "red", alpha = 0.5)
```

```python
# Split data into train and test sets
X_train = X[:40] # first 40 examples (80% of data)
y_train = y[:40]

X_test = X[40:] # last 10 examples (20% of data)
y_test = y[40:]

len(X_train), len(X_test)
```

```python
plt.scatter(X_train, y_train, color = "red", label = "Training Data")
plt.scatter(X_test, y_test, color = "blue", label = "Testing Data")
plt.legend();
```

```python
def plot_predictions(train_data=X_train, 
                     train_labels=y_train, 
                     test_data=X_test, 
                     test_labels=y_test, 
                     predictions=y_pred):
  """
  Plots training data, test data and compares predictions.
  """
  plt.figure(figsize=(10, 7))
  # Plot training data in blue
  plt.scatter(train_data, train_labels, c="b", label="Training data", alpha = 0.25)
  # Plot test data in green
  plt.scatter(test_data, test_labels, c="g", label="Testing data", alpha = 0.75)
  # Plot the predictions in red (predictions were made on the test data)
  plt.scatter(test_data, predictions, c="r", label="Predictions", alpha = 0.75)
  # Show the legend
  plt.legend();
```

```python
plot_predictions(train_data=X_train,
                 train_labels=y_train,
                 test_data=X_test,
                 test_labels=y_test,
                 predictions=y_pred)
```

```python
model.evaluate(X_test, y_test)
```
